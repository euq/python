```python
n, a, b = map(int, input().split(' '))
import math
log = int(math.log10(n))
hlist = [10 ** h for h in range(log + 1)][::-1]
nl = [i for i in range(1, n+1)]
s = list()
for nn in nl:
    print(nn)
    k = list()
    for i in hlist:
        kazu = int(nn/i)
        k.append(kazu)
        print('kazu:',kazu)
        nn -= kazu*i
        print('nn:',nn)
    print(k, sum(k))
    s.append(sum(k))

print(s)
tl = list()
for i in range(len(nl)):
    if a <= s[i] <= b:
        t = n[i]
        tl.append(t)
    
print(tl)
```
